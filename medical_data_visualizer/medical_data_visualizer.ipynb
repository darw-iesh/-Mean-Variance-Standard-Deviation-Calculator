```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
import seaborn as sns
import math
```

```python
medical_data = pd.read_csv('D:\\python\\free code camp projects\Medical Data Visualizer\medical_examination.csv')
medical_data.head()
```

```python
medical_data['overweight'] = (
    medical_data['weight'] / ((medical_data['height'] / 100) ** 2) > 25
).astype(int)
medical_data.head()
```

```python
medical_data['cholesterol'] = medical_data['cholesterol'].apply(lambda x: 0 if x == 1 else 1)
medical_data['gluc'] = medical_data['gluc'].apply(lambda x: 0 if x == 1 else 1)
medical_data.head()
```

```python
def draw_cat_plot():
    # 1. Create DataFrame for cat plot
    medical_data_cat = pd.melt(
        medical_data,
        id_vars=["cardio"],
        value_vars=[
            "active",
            "alco",
            "cholesterol",
            "gluc",
            "smoke",
            "overweight"
        ]
    )

    # 2. Group and count values
    medical_data_cat = medical_data_cat.value_counts().reset_index(name="total")

    # 3. Rename columns (IMPORTANT for catplot)
    medical_data_cat = medical_data_cat.rename(
        columns={"variable": "variable", "value": "value"}
    )

    # 4. Draw plot
    fig = sns.catplot(
        data=medical_data_cat,
        x="variable",
        y="total",
        hue="value",
        col="cardio",
        kind="bar"
    )

    fig.set_axis_labels("variable", "total")
    fig.set_titles("cardio = {col_name}")

    return fig
```

```python
def draw_heat_map():
    # Clean data
    medical_data_heat = medical_data[
        (medical_data["ap_lo"] <= medical_data["ap_hi"]) &
        (medical_data["height"] >= medical_data["height"].quantile(0.025)) &
        (medical_data["height"] <= medical_data["height"].quantile(0.975)) &
        (medical_data["weight"] >= medical_data["weight"].quantile(0.025)) &
        (medical_data["weight"] <= medical_data["weight"].quantile(0.975))
    ]

    # Correlation matrix
    corr = medical_data_heat.corr()

    # Mask upper triangle
    mask = np.triu(np.ones_like(corr, dtype=bool))

    # Figure
    fig, ax = plt.subplots(figsize=(12, 8))

    # Heatmap
    sns.heatmap(
        corr,
        mask=mask,
        annot=True,
        fmt=".1f",
        center=0,
        square=True,
        linewidths=0.5
    )

    return fig
```
