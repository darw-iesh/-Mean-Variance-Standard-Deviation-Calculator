```python
avg_age_men = df.loc[df['sex'] == ' Male', 'age'].mean()

avg_age_men.round(1)
```

```python
percentage_bachelors = (
    (df['education'] == ' Bachelors').mean() * 100
)

percentage_bachelors
```

```python
higher_education = df['education'].isin([' Bachelors', ' Masters', ' Doctorate'])

higher_education_rich = df[higher_education & (df['salary'] == ' >50K')]

percentage_higher_education_rich = (
    len(higher_education_rich) / len(df[higher_education]) * 100
)


percentage_higher_education_rich
```

```python
without_education = ~df['education'].isin(
    [' Bachelors', ' Masters', ' Doctorate']
)

without_education_rich = df[
    without_education & (df['salary'] == ' >50K')
]

percentage_without_education_rich = (
    len(without_education_rich) / len(df[without_education]) * 100
)

print(percentage_without_education_rich)
```

```python
# Minimum number of work hours
min_work_hours = df['hours_per_week'].min()
min_work_hours
```

```python
# People who work minimum hours
num_min_workers = df[df['hours_per_week'] == min_work_hours]

# Among them, those earning >50K
rich_min_workers = num_min_workers[num_min_workers['salary'] == ' >50K']

# Percentage
rich_percentage = (
    len(rich_min_workers) / len(num_min_workers) * 100
)


rich_percentage
```

```python
salary_morethan_50=df.loc[df['salary'] == ' >50K']
country_counts = salary_morethan_50['native_country'].value_counts()
number_count=country_counts.idxmax()
```

```python
(country_counts[number_count] / len(df['native_country']) * 100).round(1)
```
